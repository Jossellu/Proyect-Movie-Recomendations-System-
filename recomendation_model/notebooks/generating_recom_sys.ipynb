# Features from df to get similarity vector
which are director, cast, ,description, genres_reduction 

```python
df.columns
```

```python
features= ['director','cast','description','genres_reduction']
```

```python
df.iloc[:,0]
```

Tag column:

```python
new_df = pd.DataFrame({'title':df['title'],
                       'tags':''})
for feature in features:
    new_df['tags'] = new_df['tags'] + ' ' + df[feature] 
    if feature == features[-1]:
        new_df['tags'] = new_df['tags'] + df[feature]
    
```

```python
new_df
```

```python
new_df.to_json('../../recomendation_model/data/proccesed_data/df_to_similarity',orient='records',index=False)
```

# NLP: Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=7942 , stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags'].values.astype('U'))
```

```python
vector.shape
```

# similarity with cosine_similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity  = cosine_similarity(vector)
```

```python
similarity.shape
```

# functions to get index from title and revserse case

```python
def index_from_title(title:str): 
    return df[df['title'] == title].index[0]

def title_from_index(list_movies_indexes:list):
    for i in list_movies_indexes:
        print(df.iloc[i,0:10])
```

```python
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda idx:idx[1])[:5] #gets major similarity for a title and its idx in dataframe
```

```python
def get_recomendation(title:str, show_count:int):
    index = index_from_title(title)
    idx_similarity = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda idx:idx[1])[1:show_count+1] #gets major similarity for a title and its idx in dataframe
    movies_indexes = [elem[0] for elem in idx_similarity]
    title_from_index(movies_indexes)
```
