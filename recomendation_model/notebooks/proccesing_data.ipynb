```python
diccionario_limpio = {k.strip(): v for k, v in dicc.items() if k.strip() != ''}
diccionario_limpio
```

```python
#we generate new column genres by reducing original genres
new_generes_column = pd.Series()
for row in raw_data['genres']:
    new_row = ''
    for genre in row.split(','):    
        new_genre = genre.strip().replace(genre.strip(),diccionario_limpio[genre.strip()]) + ','
        new_row = new_row + str(new_genre)
    new_generes_column = pd.concat([new_generes_column,pd.Series(new_row)])

#delete comma at the end of each row
genres_column = pd.Series()
for row in new_generes_column:
    new_row = row[:-1]
    genres_column = pd.concat([genres_column,pd.Series(new_row)])
genres_column
```

# Short dummies
We choose classify short series ->  series_duration <2   
We choose classify short movies  ->  movies_duration < 80

```python
def only_number(string):
    patron = r'[^\d.]'
    numeros = re.sub(patron,'',string)
    return numeros
```

```python
proccesed_df['duration'] = proccesed_df['duration'].apply(only_number)
proccesed_df['duration'] = proccesed_df['duration'].astype('float')
```

```python
proccesed_df['short'] = None
```

```python
proccesed_df['short'] = 0
proccesed_df['short'] = proccesed_df.apply(lambda x:(1 if (x['type'] == 'TV Show') and (x['duration'] <2)else x['short']),axis=1)
proccesed_df['short'] = proccesed_df.apply(lambda x:(1 if (x['type'] == 'Movie') and (x['duration'] <80) else x['short']),axis=1)
```
