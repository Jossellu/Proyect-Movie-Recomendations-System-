```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/raw_data/afirnation_data.csv')
df
```

```python
from module_pro import generate_model, generate_df #slef modulo
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bowler_model = CountVectorizer() #generamos modelo de lematization
```

```python
full_data = (' ').join(df['text'].tolist())
from unidecode import unidecode
import re
full_data = unidecode(full_data)
full_data = full_data.lower()
full_data
```

```python
def quitar_no_letras_con_espacios(texto):
    return re.sub(r'[^a-zA-Z\s]', '', texto)

full_data_clean = quitar_no_letras_con_espacios(full_data)
full_data_clean
```

```python
from nltk import word_tokenize
import nltk
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('spanish')
```

```python
def stemizar(tokens):
    stemms = []
    for token in tokens:
        stemms.append(stemmer.stem(token))
    return stemms
```

```python
tok = word_tokenize(full_data_clean)
lem = stemizar(tok)
print(tok,lem,sep='*****************\n')
```

```python
bowler_model.fit(lem)
bowler_model.vocabulary_ #vocabulario de lematization de toda la data
```

```python
bowler_model.get_feature_names_out()
```

```python
df
```

```python
def limpia_texto(text):
    texto_limpio = re.sub(r'[^a-zA-Z\s]', '', text)
    sin_acentos = unidecode(texto_limpio)
    minusculas = sin_acentos.lower()
    return minusculas
```

```python
df
```

```python
df['text_clean'] = df.apply(lambda x: (limpia_texto(x['text'])),axis=1)
df['tokens'] = df.apply(lambda x: (word_tokenize(x['text_clean'])),axis=1)
df['lemms'] = df.apply(lambda x: (stemizar(x['tokens'])),axis=1)
```

```python
matrices_list = []
for lem in df['lemms']:
    matrices = bowler_model.transform(lem).toarray()
    no_matrices = len(matrices)
    shape_matrix = matrices[0].shape
    sum_matrices = np.zeros(shape_matrix)
    for matrix in matrices:
        sum_matrices = sum_matrices + matrix
    matrices_list.append(sum_matrices)
matrices_list
```

```python
df
```

```python
df['matrices'] = matrices_list
df
```

```python
df_proccesed = df[['text','matrices','afirmacion']]
df_proccesed
```

```python
df_proccesed['afirmacion']=df_proccesed['afirmacion'].astype('float32')
```

#Construimos el modelo naive steivs

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
x_data = df['matrices']
y_data = np.asarray(df['afirmacion'],dtype='float32')
```

```python
x_data.shape, y_data.shape
```

```python
x_data[0], y_data[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data,y_data,random_state=42,test_size=0.3)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
clasificator_model = MultinomialNB()
clasificator_model.fit(x_train.tolist(),y_train)
```

```python
#metricas
y_predict = clasificator_model.predict(x_test.tolist())
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred=y_predict))
print(accuracy_score(y_test,y_pred=y_predict))
```

```python
import joblib
joblib.dump(clasificator_model,'../models/afirmation_model.joblib')
```

```python
def prediccion(modelo_prediccion):
    texto = input(f'you: ')
    texto = limpia_texto(texto)
    tokens = word_tokenize(texto)
    leems = stemizar(tokens)
    matriz = bowler_model.transform(leems).toarray()
    predict = modelo_prediccion.predict(matriz)
    #print(modelo_prediccion.predict_proba(matriz))
    print(predict)
```

```python
while True:
    prediccion(clasificator_model)
```
