```python
# @codificacion de un texto simple
#paso 1 importar modulo bower para obtener data
from concverter2_0 import bower
#paso 2 obtener tokens,stops y lemms
tokens, stops, lemms = bower().text(text='!Hola!, como has estado? yo bien muchas gracias')
#paso 3 crear modelo bower con sklearn y ajustarlo con tus lemms de todo tu data
from sklearn.feature_extraction.text import CountVectorizer
bower_model = CountVectorizer()
bower_model.fit(lemms)
#paso 4 transoformar tus leems por elemento de tu data con el modelo ajustado 
coded = bower_model.transform(lemms)
#paso 5 observar tus features para codificar texto
bower_model.get_feature_names_out()
#paso 6 convertirlo en array para observarlo en forma de matriz
coded = coded.toarray()
```

```python
print('tokens: ', tokens)
print('no stop words: ', tokens)
print('lemmatized : ', lemms)
print('features to code: ', bower_model.get_feature_names_out())
print('texto codificado: ', coded)
```
